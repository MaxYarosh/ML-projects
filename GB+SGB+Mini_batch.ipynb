```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function and its derivative
def f(x):
    return (x - 3) ** 2 + 4

def grad(x):
    return 2 * (x - 3)
```

```python
# GD
def gradient_descent(lr=0.1, epochs=20):
    x = 10.0  # старт
    history = [x]
    for _ in range(epochs):
        x -= lr * grad(x)
        history.append(x)
    return history
```

```python
# SGD
def stochastic_gradient_descent(lr=0.1, epochs=20):
    x = 10.0
    history = [x]
    for _ in range(epochs):
        noise = np.random.normal(0, 1)
        x -= lr * (grad(x) + noise*0.3)
        history.append(x)
    return history
```

```python
# Mini-batch
def mini_batch_gradient_descent(lr=0.1, epochs=20, batch_size=5):
    x = 10.0
    history = [x]
    for _ in range(epochs):
        for _ in range(batch_size):
            noise = np.random.normal(0, 0.3)
            x -= lr * (grad(x) + noise)
        history.append(x)
    return history
```

```python
gd_path = gradient_descent()
sgd_path = stochastic_gradient_descent()
mbgd_path = mini_batch_gradient_descent()

x_vals = np.linspace(-2, 10, 200)
plt.plot(x_vals, f(x_vals), label="Function f(x)", color="black")

plt.scatter(gd_path, [f(x) for x in gd_path], color="blue", label="Gradient Descent", marker="o")
plt.scatter(sgd_path, [f(x) for x in sgd_path], color="red", label="SGD", marker="x")
plt.scatter(mbgd_path, [f(x) for x in mbgd_path], color="green", label="Mini-batch GD", marker="s")

plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Comparison GD, SGD and Mini-batch GD")
plt.show()
```
